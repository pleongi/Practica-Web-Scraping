```python
#referencia request y webscraping:
#https://code.tutsplus.com/tutorials/using-the-requests-module-in-python--cms-28204
#https://code.tutsplus.com/es/tutorials/scraping-webpages-in-python-with-beautiful-soup-the-basics--cms-28211
# 1er instalar requests para importar la web, tambien nos permitar hacer el post request y despues installar beautifulsoup
# es probable que necesitemos panda para crear el csv.
import requests
import csv
from bs4 import BeautifulSoup
#seleccionamos la web con los datos solo de barcelona ,que son los que queremos extraer
req = requests.get('https://lap.uab.cat/aerobiologia/es/forecast/barcelona')
#crea la variable polenesB que contiene el contenido de la web a estudio, solo de barcelona
polenesB=req.content
```

```python
#ahora hay que modelizar y procesamos la fuente a bautifulsoup
soup= BeautifulSoup(polenesB, 'lxml')
# exploramos la web con prettify que nos da la informacion en nidos
#print(soup.prettify()) 
#observamos  que los datos que nos interesan estan en table.
table=soup.find('table')
print (table)
```

```python
# esta sintaxis es de Paula, tengo problemas para ejecutarla.
import pandas as pd
new_table = pd.DataFrame(columns=range(0,78), index = [0])
row_marker = 0
for row in table.find_all('tr'):
  column_marker = 0
  columns = row.find_all('td')
  for column in columns:
    
    print("row:",row_marker,column_marker)
    print("col",column.get_text())
    new_table.iat[row_marker,column_marker] = column.get_text()
    column_marker += 1
    
new_table
filename = "polen_barcelona_dataset.csv"
new_table.to_csv(filename)
```
