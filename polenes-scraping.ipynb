```python
#referencia request y webscraping:
#https://code.tutsplus.com/tutorials/using-the-requests-module-in-python--cms-28204
#https://code.tutsplus.com/es/tutorials/scraping-webpages-in-python-with-beautiful-soup-the-basics--cms-28211
# 1er instalar requests para importar la web, tambien nos permitar hacer el post request y despues installar beautifulsoup
import requests
req = requests.get('https://lap.uab.cat/aerobiologia/es/forecast/catalunya')
#req.content evaluo el contenido de la web.
req.content
```

```python
from bs4 import BeautifulSoup
#creo la variable src que contiene el contendio de la web a estudio
src=req.content
```

```python
#ahora hay que modelizar y procesamos la fuente a bautifulsoup
soup= BeautifulSoup(src, 'html')
```

```python
#busco las estaciones.
estacion= soup.find_all("th")
print(estacion)
print ('\n')
```

```python
polenes=soup.find_all("td")
print(polenes)
print ('\n')
```

```python
#informacion en nidos
print(soup.prettify())
```
