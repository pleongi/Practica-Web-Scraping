```
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup  

#link = "https://lap.uab.cat/aerobiologia/es/forecast/barcelona"
link="https://lap.uab.cat/aerobiologia/es/forecast/catalunya"

f = urlopen(link)
myfile = f.read()
#print(myfile)

page_soup = soup(myfile, "lxml")

#print(page_soup.prettify())
table=page_soup.find_all('table')[1]

dfs=[]
import pandas as pd
df_list = pd.read_html(str(table))
filename = "polen_dataset.csv"
df = pd.concat(df_list)  
dfs.append(df)

final_df = pd.concat(dfs)
final_df.to_csv(filename,sep=',')
final_df
```
