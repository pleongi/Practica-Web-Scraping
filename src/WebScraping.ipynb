```
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup  

link="https://lap.uab.cat/aerobiologia/es/forecast/catalunya"


from urllib.robotparser import RobotFileParser
rp = RobotFileParser()
rp.set_url("https://lap.uab.cat/robots.txt") #no hay fichero robots.txt en esta web
rp.read()
rp.can_fetch("*", link)


f = urlopen(link)
myfile = f.read()

page_soup = soup(myfile, "lxml")
table=page_soup.find_all('table')[1]

dfs=[]
import pandas as pd
df_list = pd.read_html(str(table))
filename = "polen_dataset.csv"
df = pd.concat(df_list)  
dfs.append(df)

final_df = pd.concat(dfs)

names=[]
for i,name in zip(range(len(final_df.iloc[0])),final_df.iloc[0]):
  if(i!=0):
    names.append(name)
    names.append(name)
  else:
    names.append(name)
cleanedList = [x for x in names if str(x) != 'nan']

names=[]
for name in final_df.iloc[:,0]:
    names.append(name)
      

final_df.columns = cleanedList
final_df.index = names
final_df=final_df.drop('Taxones pólenes', axis=1)
final_df=final_df.drop(['Taxones pólenes'])
final_df=final_df.drop(['Taxones esporas hongos'])

final_df.to_dense().to_csv("polen_dataset.csv", index = True, sep=',', encoding='utf-8')
final_df
```
