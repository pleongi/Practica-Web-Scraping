```
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup  

link = "https://lap.uab.cat/aerobiologia/es/forecast/barcelona"
final_link="https://lap.uab.cat/aerobiologia/es/forecast/catalunya"

f = urlopen(link)
myfile = f.read()
#print(myfile)

page_soup = soup(myfile, "lxml")

#print(page_soup.prettify())
table=page_soup.find('table')


import pandas as pd
new_table = pd.DataFrame(columns=range(0,78), index = [0])
row_marker = 0
for row in table.find_all('tr'):
  column_marker = 0
  columns = row.find_all('td')
  for column in columns:
    
    print("row:",row_marker,column_marker)
    print("col",column.get_text())
    new_table.iat[row_marker,column_marker] = column.get_text()
    column_marker += 1
    
new_table
filename = "polen_barcelona_dataset.csv"
new_table.to_csv(filename)
```
